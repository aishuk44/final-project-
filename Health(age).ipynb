#### **Introduction**
The metric I chose is "Average MD Age at Death" which kind of realtes to how healthy the neighborhood is, so depeding on the highest total average median age at death I could determine the which neiughborhood is healthiest.

```python
data = pd.read_csv("MD_death_rate.csv")
data.head(91)
```

```python

#table1= pd.DataFrame(data)
data_set = data.set_index("NEIGHBORHOOD")
data_set=data_set.drop(["BLACKdeaths", "Black MD AGE AT DEATH", "WHITEdeaths", "White MD AGE AT DEATH"], axis=1)

data_set.head(91)
```

After reading in the file, I dropped extra coloumns to make my table much clear to follow up

```python
data_set['TOTALdeaths*'].mean()
```

Since there was the neighborhoods with no records of deatgh rate, I needed to get average raw death rate.

```python
death_QM = data_set[data_set['TOTALdeaths*'] > 171 ]
death_QM.head()
```

After finding out the average death rate amoung all neighborhoods, I dropped all neighborhoods that below from average death rate. So it came down to pretty simple table

```python
death_QM.sort_values(by=['TOTAL MD AGE AT DEATH'], ascending = False)
```

```python
death_QM.plot(kind = "bar", figsize = (10, 10 ))
```
